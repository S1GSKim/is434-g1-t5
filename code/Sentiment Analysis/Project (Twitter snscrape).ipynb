### Twitter snscrape

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

Pre, During, Post sentiment analysis on covid-19
Keywords (need to do both with & without hyphen):
Vaccine,
Vaccines,
anti-vaccine,
anti-vaxxer,
anti-vax

## Pre-covid

Scraping tweets from a text search query

Using the code below, we are scraping tweets between **July 01, 2019, and October 31, 2019 (Pre-Covid)**,
with the keyword <font color=#FF0000>'vaccines'</font>.

Pulled attributes DateTime, tweet id, text, and username from the tweet objects.

```python
# Creating list to append tweet data to
tweets_list = []

# search terms
search_terms = ['anti-vaccine','anti-vaxxer','anti-vax']

for search_term in search_terms:
    query = f'{search_term} since:2019-07-01 until:2019-10-31'
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
        tweets_list.append([tweet.date, tweet.content, tweet.user.username])
    
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
tweets_df
```

```python
tweets_df.to_csv('pre_covidTweets.csv', index=False)
```

## During Covid

### Part 1: Scraping tweets from a text search query

Using the code below, we are scraping tweets between **Decemeber 01, 2020, and February 28, 2021 (During Covid)**, with the keywords <font color=#FF0000>'vaccines'</font>.

Pulled attributes DateTime, tweet id, text, and username from the tweet objects.

```python
# Creating list to append tweet data to
tweets_list2 = []

# search terms
search_terms = 'vaccines'

# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper(search_terms + ' since:2020-12-01 until:2021-02-28').get_items()):
    if i > 20000:
        break
    #declare the attributes to be returned
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.user.username])

# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

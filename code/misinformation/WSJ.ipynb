```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# users = ['ABC','FoxNews','MSNBC','NBCNews','USATODAY','WSJ','washingtonpost', 'BuzzFeed', 'voxdotcom', 'MailOnline', 'Daily_Express','Telegraph'])
import re
from newspaper import Article
users = ['WSJ']
filters = ['Covid-19 since:2020-12-01', 'until:2021-02-28']
from_filters = []
for user in users:
    from_filters.append(f'from:{user}')
filters.append(' OR '.join(from_filters))
tweets = list(sntwitter.TwitterSearchScraper(' '.join(filters)).get_items())
print(len(tweets))
tweet_list=[]
count=1
for tweet in tweets:
    tweetContent=tweet.rawContent
    try:
        url = re.search("(?P<url>https?://[^\s]+)", tweetContent).group("url")
        print(url)
        count+=1
        try:
            toi_article = Article(url, language="en") # en for English

            #To download the article
            toi_article.download()

            #To parse the article
            toi_article.parse()

            #To perform natural language processing ie..nlp
            toi_article.nlp()

            tweet_list.append([tweet.date, tweet.retweetCount, toi_article.text, tweet.user.username, toi_article.keywords])
            print("Able to Scrape article"+str(count))
        except:
            tweet_list.append([tweet.date, tweet.retweetCount, tweet.rawContent, tweet.user.username, []])
            print("Unable to Scrape article"+str(count))
    except:
        tweet_list.append([tweet.date, tweet.retweetCount, tweet.rawContent, tweet.user.username, []])
        print("Unable to Scrape article"+str(count))
df_first = pd.DataFrame(tweet_list, columns=['Datetime','retweetCount','Text', 'Username','Keywords'])
df_first.drop_duplicates(subset='Text', keep="last")
df_first.head(30)
```
